```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_au")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_Movies_FilmDetails_transformed = spark.table(f"{catalogo}.{esquema_source}.Movies_FilmDetails")
```

```python
df_transformed = df_Movies_FilmDetails_transformed.agg(
    count("*").alias("conteo"),
    max(col("user_score")).alias("max_user_score"),
    min(col("runtime_min")).alias("min_runtime")
)
df_transformed.show()
```

```python
df_transformed.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.df_Movies_FilmDetails_transformed")
```
