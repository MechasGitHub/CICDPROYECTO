```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_au")
dbutils.widgets.text("esquema_source", "bronze")
dbutils.widgets.text("esquema_sink", "silver")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_movies = spark.table("catalog_au.bronze.Movies")
df_filmdetails = spark.table("catalog_au.bronze.FilmDetails")
```

```python
df_joined = df_movies.join(df_filmdetails, df_movies.movie_id == df_filmdetails.movie_id, "inner")
```

```python
df_transformed = df_joined.withColumn(
    "movie_title_details",
    F.concat_ws(" - ", df_movies.title, df_filmdetails.details)
)
```

```python
df_transformed.select(
    "movie_id", "title", "details", "movie_title_details"
).write.mode("overwrite").saveAsTable("catalog_au.bronze.Movies_FilmDetails")
```
